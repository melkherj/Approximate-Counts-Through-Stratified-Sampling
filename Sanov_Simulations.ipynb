```python
# Choice of alpha
p = np.arange(0.02,0.98,0.01)

alpha = map(choose_alpha,p)
plot(p,map(choose_alpha,p),label='optimal alpha')
plot(p,[2]*len(p),label='hoeffding')
#plot(p,1/(3.*p),label='chernoff') -- doesn't hold
ylim(0,20)
legend()
```

```python
p = 0.5
alpha = choose_alpha(p)
qs = np.arange(p,0.99,0.001)
kl = np.array([q*log(q/p) + (1-q)*log((1-q)/(1-p)) for q in qs])
#chernoff = np.array([1/(3.*p)*(q-p)**2 for q in qs]) -- doesn't hold
dkl = np.array([(log(q/(1-q)) - log(p/(1-p))) for q in qs])
d2kl = np.array([(1/(q*(1-q))) for q in qs])
f = np.array([alpha*(q-p)**2 for q in qs])
df = np.array([2*alpha*(q-p) for q in qs])
d2f = np.array([2*alpha for q in qs])
print min(kl  -  f)
print 'Worst ratio kl/quadratic: %.3f'%(kl[-1] / f[-1])
```

```python
plot(qs,f,label='f')
plot(qs,kl,label='kl')
#plot(qs,chernoff,label='chernoff') -- doesn't hold
title('kl vs. best quadratic approximation')
legend(bbox_to_anchor=[1.5, 0.5]); show()

plot(qs,df,label='df')
plot(qs,dkl,label='dkl')
title('dkl vs. dquadratic')
legend(bbox_to_anchor=[1.5, 0.5]); show()

plot(qs,d2f,label='d2f')
plot(qs,d2kl,label='d2kl')
title('d2kl vs. d2quadratic')
legend(bbox_to_anchor=[1.5, 0.5]); show()

print 'Worst ratio kl/quadratic: %.3f'%(kl[-1] / f[-1])
```
