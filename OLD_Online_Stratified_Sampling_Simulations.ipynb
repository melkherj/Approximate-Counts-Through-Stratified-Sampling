### Uniform Empirical Bernstein Bound ###
This is Akshay's new result, that gets rid of the $log(t)$ term in the uniform empirical bernstein bound.  

Let $C_\delta = 8log\left(\frac{2*\sqrt{2}}{\delta}\right)$.  

Then with probability $1-\delta$, for every $t>C_\delta$ simultaneously, 

$$\left|\hat{\mu}- \mu\right| \leq \frac{\frac{5}{6}C_\delta}{t-C_\delta}+\hat{\sigma_t}\sqrt{\frac{C_\delta}{t-C_\delta}}$$

### Cumulative Computations, and other Helper Functions ###
These include:

* $\hat{p_t}$: Cumulative average
* $\hat{\sigma_t}$: Cumulative standard deviation
* $\epsilon_t$ (bernstein): Cumulative error bounds on the $\hat{p_t}$ estimator, that hold uniformly with probability $1-\delta$
* prior_term/empirical_term: helper functions that compute the first and second terms of the error above

```python
def sample_dist(dist, samples=10):
    ''' Given a discrete distribution, eg [0.1,0.2,0.7], generate <samples>
        samples from the distribution, eg [0,1,2,1,2,2,2,1,0] '''
    cumulative = cumsum(dist)
    return [nonzero(cumulative > r)[0][0] for r in rand(samples)]

def sample_bernouli(p,samples=10):
    dist = array([1-p,p])
    return sample_dist(dist, samples=samples)

def cum_average(samples):
    ''' Given a list of 0/1 binary iid samples from a discrete distribution, 
        generate a sequence of empirical averages of these samples'''
    n = len(samples)
    return np.cumsum(samples)/np.arange(1.0,n+1)
    
def cum_std(samples):
    ''' Given a list of 0/1 binary iid samples from a discrete distribution, 
        generate a sequence of empirical standard deviation estimations'''
    p = cum_average(samples)
    return np.sqrt(p*(1-p))

def C_coefficient(delta):
    return 8*log(2*sqrt(2)/delta)

def drop_max(eps,t0):
    ''' Given a numpy array <eps>, that gives the some error bound over time, 
        and a time t0 (with time indexed starting at 1), use a trivial bound
        for cases when t <= t0 or eps[t] > 1'''
    eps[:int(t0)+1] = 1
    eps[eps > 1] = 1
    return eps

def prior_term(n, delta):
    ''' Gives the error bound for the c/t term of the empirical bernstein bound '''
    C = C_coefficient(delta)
    T = np.arange(1.0,n+1)
    # The bound error at each timestep, as a numpy array
    eps = (5./6)*C/(T-C)
    eps = drop_max(eps,C)
    return eps

def empirical_term(samples, delta):
    ''' The second term of the empirical bernstein bound above, including the empirical variance '''
    n = len(samples)
    C = C_coefficient(delta)
    T = np.arange(1.0,n+1)
    s = cum_std(samples)
    eps = s*sqrt(C/(T-C))
    eps = drop_max(eps,C)
    return eps

def bernstein(samples, delta):
    n = len(samples)
    C = C_coefficient(delta)
    eps = prior_term(n,delta) + empirical_term(samples,delta)
    eps = drop_max(eps,C)
    return eps

def np_first(np_arr, min_index=0):
    ''' Get the index of the first true element of a numpy array.  
        Returns -1 if there is no true element'''
    try:
        z = np.nonzero(np_arr)[0]
        return int(z[z >= min_index][0])
    except IndexError:
        return -1
    
def stoptime(eps, eps_final,min_time=0):
    ''' Get the first time t where eps[t] <= eps_final '''
    return np_first(eps <= eps_final, min_index=min_time)

def prior_stoptime(delta,eps):
    ''' Closed-form solution for the stopping time for the prior term '''
    C = C_coefficient(delta)
    return C/eps+C

def bernstein_stoptime(p,delta,t_max,eps_final):
    ''' Generates samples from a bernouli distribution with the given p.  
        Returns the number of iterations needed before the UEB bound can be used
        to get the actual p, to within <= eps_final error '''
    samples = sample_bernouli(p,samples=t_max)
    eps = bernstein(samples,delta)
    return stoptime(eps,eps_final)

def plot_errs(samples, delta, title_str=None):
    ''' Given iid samples from some bernouli distribution, plot the first and second error terms 
        in the empirical bernstein bound, along with the sum of these errors, 
        as a function of the number of examples processed '''
    n = len(samples)
    h = prior_term(n,delta)
    e = empirical_term(samples, delta)
    eps = bernstein(samples, delta)
    ax = gca()
    plot(h,label='prior')
    plot(e,label='empirical')
    plot(eps,label='bernstein=prior+empirical')
    xlim([C_coefficient(delta),n])
    ax.set_yscale('log')
    xlabel('Number of examples (t)')
    ylabel('Error bound')
    if not title_str is None:
        title(title_str)
    legend()
    show()
```

```python
def stratification_improvement(ns, ps, delta, eps_final, alpha=1.0):
    ''' For a partition of the data into two sets, compare the performance of running 
        uniform and stratified sampling.  This sampling is to approximate the total proportion of 
        queries matching some criteria with accuracy eps_final and with probability 1-delta'''
    print 'ns: %d/%d'%tuple(ns)
    print 'ps: %.2f/%.2f'%tuple(ps)
    # The number of items in the entire table
    N = sum(ns)
    # Total samples to generate
    t0 = 100000
    stratified_samples = [sample_bernouli(p,int(t0*float(n)/N)) for n,p in zip(ns,ps)]
    all_samples = [s for samples in stratified_samples for s in samples]
    random.shuffle(all_samples)
    uniform_stoptime = stoptime(bernstein(all_samples,delta),eps_final)
    print 'Uniform sampling stoptime: %d'%uniform_stoptime
    #eps0 = alpha*eps_final
    eps0 = N/float(ns[0])*eps_final/2.
    eps1 = N/float(ns[1])*eps_final/2.
    eps0 = alpha*eps0
    eps1 = (N*eps_final-ns[0]*eps0)/ns[1]
    assert (eps0*ns[0]+eps1*ns[1])/N == eps_final
    t0 = stoptime(bernstein(stratified_samples[0],delta/2), eps0)
    t1 = stoptime(bernstein(stratified_samples[1],delta/2), eps1)
    print 'Partition 0/1 stoptimes: %d/%d'%(t0,t1)
    print 'Max/total stoptime: %d/%d'%(max(t0,t1), t0+t1)
    print 'Estimated ps for partitions 0/1: %.3f/%.3f'%(  mean(stratified_samples[0][:t0]), 
                                  mean(stratified_samples[1][:t1])  )
    print 'Max time improvement ratio: %.3f\n\n'%(float(uniform_stoptime)/max(t0,t1))
```

```python
print "'delta' gives the probability the final confidence interval won't hold"
print "'epsilon_final' gives the width of the confidence interval on the final approximate count"
print "'ns' the number of items (table rows) in each partition of the table"
print "'ps' defines a bernouli distribution for each partition, giving the proportion of each partition \n \
 that matches some particular query Q. \n \
 Note: p does not necessarily need to sum to 1"

delta = 0.05
eps_final = 0.01

ns = np.array([100,10])
ps = array([0.01,0.99]) # highly non-uniform
assert len(ns) == len(ps) #the number of partitions
print "For the following, we've fixed delta=0.05, final accuracy = epsilon = 0.01\n\n"

print "### Using suboptimal, local allocation of epsilon to partitions ###\n"
stratification_improvement(array([10,10]), array([0.01,0.99]), delta, eps_final)
stratification_improvement(array([1000,10]), array([0.01,0.99]), delta, eps_final)
stratification_improvement(array([10,10]), array([0.00,0.99]), delta, eps_final)
stratification_improvement(array([10,10]), array([0.1,0.1]), delta, eps_final)

print "### Using approximately optimal (for max time) allocation of epsilon to partitions ###\n"
stratification_improvement(array([10,10]), array([0.01,0.99]), delta, eps_final, alpha=1.0)
stratification_improvement(array([1000,10]), array([0.01,0.99]), delta, eps_final, alpha=1.85)
stratification_improvement(array([10,10]), array([0.00,0.99]), delta, eps_final, alpha=0.6)
stratification_improvement(array([10,10]), array([0.1,0.1]), delta, eps_final, alpha=1.0)
```

### Power-Law for Crossover Times ###
The following is a pretty trivial result.  There is a power law relating $p$, the proportion of a sample matching a query, 
and the time when the prior/empirical terms of the bernstein bound we've been discussing cross.  

Finding the formula should require a bit of algebra.  Not very much.  

```python
ps, cross_ts, _ = zip(*crossovers)
ax = gca()
ax.set_yscale('log')
ax.set_xscale('log')
plot(ps,cross_ts)
xlabel('p: proportion of samples matching table query')
ylabel('crossover time')
show()
```
